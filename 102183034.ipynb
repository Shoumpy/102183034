```python
import numpy as np
import pandas as pd
import math as m
import sys

if len(sys.argv) < 4:
  print("not enough arguments")
  sys.exit()

filename = sys.argv[1]
weight = sys.argv[2]
impact = sys.argv[3]

#weights and impacts should be seperated by a comma
weight = list(map(float, weight.split(",")))
impact = list(map(str, impact.split(",")))

dataset = pd.read_csv(102183034-data.csv.csv)
data = dataset.iloc[:,1:].values.astype(float)
(r,c)=data.shape

s = sum(weight)
if len(weight)<c:
  print('insufficient weights')
  sys.exit()

for i in range(c):
  weight[i]/=s

a = [0]*(c)

for i in range(r):
  for j in range(c):
    a[j] = a[j] + (data[i][j]*data[i][j])

for j in range(c):
  a[j]=m.sqrt(a[j])

for i in range(r):
  for j in range(c):
    data[i][j]/=a[j]
    data[i][j]*=weight[j]

print(data)

ideal_positive = np.amax(data,axis=0)
ideal_negative = np.amin(data,axis=0)
for i in range(len(impact)):
  if(impact[i]=='-'):
    temp = ideal_positive[i]
    ideal_positive[i] = ideal_negative[i]
    ideal_negative[i] = temp

dist_pos = list()
dist_neg = list()

for i in range(r):
  s=0
  for j in range(c):
    s+=pow((data[i][j]-ideal_positive[j]),2)

  dist_pos.append(float(pow(s,0.5)))

for i in range(r):
  s=0
  for j in range(c):
    s+=pow((data[i][j]-ideal_negative[j]),2)
  dist_neg.append(float(pow(s,0.5)))

performance_score = dict()

for i in range(r):
  performance_score[i+1] = dist_neg[i]/(dist_pos[i]+dist_neg[i])

a=list(performance_score.values())
b=sorted(list(performance_score.values()), reverse = True)

rank = dict()

for i in range(len(a)):
  rank[(b.index(a[i])+1)] = a[i]
  b[b.index(a[i])] = -b[b.index(a[i])]

row = list(i+1 for i in range(len(b)))
a = list(rank.values())
rr = list(rank.keys())

out={'Row_no':row, 'Performance_score' : a, 'Rank' : rr}
output=pd.Dataframe(out)
print(output)





```
